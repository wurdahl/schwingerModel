```python
import numpy as np
from typing import NamedTuple
```

```python
class bilinear(NamedTuple):
    aq: str
    q: str
    gamma: np.ndarray
    D: int

GAMMA5 = np.array([[1j,0],[0,-1j]])
```

```python
pion1 = bilinear(aq="u",q="d",gamma=GAMMA5, D=0)
pion2 = bilinear(aq="d",q="u",gamma=GAMMA5, D=0)
```

```python
source = [pion1]
sink = [pion2]
```

```python
import itertools
```

```python
start = [pion1, pion2]

print(start)
print()

for perm in itertools.permutations(start):
    print(perm)
    for i in perm:
        for j in start:
            print(i.aq == j.q)

    print()
```
